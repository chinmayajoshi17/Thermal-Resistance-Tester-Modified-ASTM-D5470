```python
def statisticsOut(dataName):
    means = []
    stDevs = []

    for i in range(dataName.shape[1]):
        np.mean(dataName[i])
        np.std(dataName[i])

        means.append(np.mean(dataName[i]))
        stDevs.append(np.std(dataName[i]))

    #print(means)
    #print(stDevs)

    return means, stDevs
```

```python
statisticsData1 = np.array(statisticsOut(TRTdata1))
statisticsData2 = np.array(statisticsOut(TRTdata2))
statisticsData3 = np.array(statisticsOut(TRTdata3))
statisticsData4 = np.array(statisticsOut(TRTdata4))



def linearRegressionData(sampleThicknesses, yAxisValues):
    regressionLine = np.polyfit(sampleThicknesses, yAxisValues, 1)
    dataOut = np.polyval(regressionLine, sampleThicknesses)

    return regressionLine, dataOut

fitLine1 = linearRegressionData(sampleThicknessesTiG2, statisticsData1[0])
fitLine2 = linearRegressionData(sampleThicknessesTiG2, statisticsData2[0])
fitLine3 = linearRegressionData(sampleThicknessesPG, statisticsData3[0])
fitLine4 = linearRegressionData(sampleThicknessesPG, statisticsData4[0])
```

```python
def plotWithErrorBars(statisticsData, errorData, sampleThicknesses, fitLine, Title):

    statisticsData = np.array(statisticsData)
    errorData = np.array(errorData)
    sampleThicknesses = np.array(sampleThicknesses)


    plt.figure()
    #plt.scatter(sampleThicknesses, statisticsData[0], label="PG With TIM")
    plt.errorbar(sampleThicknesses*1000, statisticsData[0]*1000,yerr = errorData*1000, capsize= 4, fmt='bo')
    plt.plot(sampleThicknesses*1000, fitLine[1]*1000, 'r-')

    plt.title(Title, fontsize= 14)
    plt.xlabel("Thickness (mm)", fontsize= 14)
    plt.ylabel("Thermal Resistance ($m^{2}K/kW$)", fontsize= 14)
    plt.xticks(fontsize= 14)
    plt.yticks(fontsize= 14)
    plt.show()

    return 0


def plotCombinePlots(statisticsDataTIM, statisticsDataNoTIM, errorDataTIM, errorDataNoTIM, sampleThicknesses, fitLineTIM, fitLineNoTIM, Title):

    sampleThicknesses = np.array(sampleThicknesses)
    statisticsDataTIM = np.array(statisticsDataTIM)
    errorDataTIM = np.array(errorDataTIM)
    statisticsDataNoTIM = np.array(statisticsDataNoTIM)
    errorDataNoTIM = np.array(errorDataNoTIM)

    plt.figure()
    #plt.scatter(sampleThicknesses, statisticsDataTIM[0], label="With TIM")
    plt.errorbar(sampleThicknesses*1000, statisticsDataTIM[0]*1000,yerr = errorDataTIM*1000, capsize= 4, fmt='bo', label='With TIM')
    plt.plot(sampleThicknesses*1000, fitLineTIM[1]*1000, 'b-')

    #plt.scatter(sampleThicknesses, statisticsDataNoTIM[0], label="Without TIM")
    plt.errorbar(sampleThicknesses*1000, statisticsDataNoTIM[0]*1000,yerr = errorDataNoTIM*1000, capsize= 4, fmt='ro', label='Without TIM')
    plt.plot(sampleThicknesses*1000, fitLineNoTIM[1]*1000, 'r-')
    plt.legend()
    plt.grid(True)
    plt.suptitle(Title, fontsize= 14, y = -0.025)
    plt.xlabel("Thickness (mm)", fontsize= 14)
    plt.ylabel("Thermal Resistance ($m^{2}K/kW$)", fontsize= 14)
    plt.xticks(fontsize= 14)
    plt.yticks(fontsize= 14)
    plt.show()

    return 0
```

```python
plotWithErrorBars(statisticsData1, uRth[0], sampleThicknessesTiG2, fitLine1, "TiG2 With TIM")
plotWithErrorBars(statisticsData2, uRth[1], sampleThicknessesTiG2, fitLine2, "TiG2 Without TIM")
plotWithErrorBars(statisticsData3, uRth[2], sampleThicknessesPG, fitLine3, "PG With TIM")
plotWithErrorBars(statisticsData4, uRth[3], sampleThicknessesPG, fitLine4, "PG Without TIM")

```

```python
plotCombinePlots(statisticsData1, statisticsData2, uRth[0], uRth[1], sampleThicknessesTiG2, fitLine1, fitLine2, "(a)")
plotCombinePlots(statisticsData3, statisticsData4, uRth[2], uRth[3], sampleThicknessesPG, fitLine3, fitLine4, "(b)")

print(fitLine1[0], fitLine2[0])
print(fitLine3[0], fitLine4[0])
```
